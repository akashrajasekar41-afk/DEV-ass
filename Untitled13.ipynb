1Ô∏è‚É£ LINE PLOT

```python
import matplotlib.pyplot as plt

experience = [1, 2, 3, 4, 5]
salary = [25000, 30000, 40000, 50000, 65000]

plt.plot(experience, salary)
plt.title("Line Plot: Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

üìù Explanation

A line plot is used to show trend or change over time.
Here, it shows how salary increases as experience increases.

2Ô∏è‚É£ BAR PLOT

```python
import matplotlib.pyplot as plt

departments = ["IT", "HR", "Sales", "Finance"]
employees = [25, 10, 30, 15]

plt.bar(departments, employees)
plt.title("Bar Plot: Department-wise Employees")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.show()
```

üìù Explanation

A bar plot is used to compare different categories.
Each bar represents the number of employees in a department.

3Ô∏è‚É£ HISTOGRAM

```python
import matplotlib.pyplot as plt

ages = [18, 22, 25, 30, 35, 40, 45, 50, 55]

plt.hist(ages, bins=5)
plt.title("Histogram: Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

üìù Explanation

A histogram shows the frequency distribution of numerical data.
It helps identify how data is spread across intervals.

4Ô∏è‚É£ SCATTER PLOT

```python
import matplotlib.pyplot as plt

experience = [1, 2, 3, 4, 5]
salary = [25000, 30000, 40000, 50000, 65000]

plt.scatter(experience, salary)
plt.title("Scatter Plot: Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

5Ô∏è‚É£ BOX PLOT

```python
import matplotlib.pyplot as plt

salary = [25000, 30000, 40000, 50000, 65000, 100000]

plt.boxplot(salary)
plt.title("Box Plot: Salary Distribution")
plt.show()
```

üìù Explanation

A box plot shows minimum, maximum, median, quartiles, and outliers.
It is mainly used for outlier detection.

6Ô∏è‚É£ PIE CHART

```python
import matplotlib.pyplot as plt

departments = ["IT", "HR", "Sales", "Finance"]
employees = [25, 10, 30, 15]

plt.pie(employees, labels=departments, autopct="%1.1f%%")
plt.title("Pie Chart: Department Distribution")
plt.show()
```

üìù Explanation

A pie chart represents percentage contribution of categories.
Each slice shows the proportion of employees per department.

7Ô∏è‚É£ HEATMAP

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Age": [22, 25, 30, 35, 40],
    "Salary": [30000, 35000, 50000, 60000, 75000],
    "Experience": [1, 2, 5, 8, 12]
}

df = pd.DataFrame(data)

sns.heatmap(df.corr(), annot=True)
plt.title("Heatmap: Correlation Matrix")
plt.show()
```


üìù Explanation
A heatmap visualizes correlation between numerical features.
Darker colors indicate stronger relationships.



8Ô∏è‚É£ COUNT PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

departments = ["IT", "HR", "IT", "Sales", "HR", "IT", "Finance"]

sns.countplot(x=departments)
plt.title("Count Plot: Department Count")
plt.show()
```

üìù Explanation

A count plot counts occurrences of categorical values.
It is mainly used in Exploratory Data Analysis (EDA).

9Ô∏è‚É£ REGRESSION PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Experience": [1, 2, 3, 4, 5],
    "Salary": [25000, 30000, 40000, 50000, 65000]
}

df = pd.DataFrame(data)

sns.regplot(x="Experience", y="Salary", data=df)
plt.title("Regression Plot: Experience vs Salary")
plt.show()
```

üìù Explanation

A regression plot shows data points with a best-fit line.
It helps understand trends and make predictions.

üîü KDE PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

salary = [25000, 30000, 40000, 50000, 65000, 80000]

sns.kdeplot(salary, fill=True)
plt.title("KDE Plot: Salary Distribution")
plt.show()
```

üìù Explanation

A KDE plot shows the probability density of data.
It is a smooth version of a histogram.
